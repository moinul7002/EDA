```python
# Visualizing Relationships
```

```python
fig, ax = plt.subplots(figsize=(12, 8))

plt.scatter(housing_data['total_rooms'], housing_data['median_house_value']) #scattered plot to visualize the relationship between total rooms and median house price

plt.xlabel('Total rooms')
plt.ylabel('Median house value')
```

```python
fig, ax = plt.subplots(figsize=(12, 8))

plt.scatter(housing_data['housing_median_age'], housing_data['median_house_value']) #scattered plot to visualize the relationship between total rooms and median house price

plt.xlabel('Median age')
plt.ylabel('Median house value')
```

```python
fig, ax = plt.subplots(figsize=(12, 8))

plt.scatter(housing_data['median_income']* 10000, housing_data['median_house_value']) #scattered plot to visualize the relationship between total rooms and median house price

plt.xlabel('Median income')
plt.ylabel('Median house value')
```

```python
housing_data_corr = housing_data.corr() #correlation with each features and target
housing_data_corr

#same direction - positive correlated
#opposite direction - negative correlated
```

```python
fig, ax = plt.subplots(figsize=(12,10))

sns.heatmap(housing_data_corr, annot = True) #correlation best visulaized using heatmap (seaborn library)
```
